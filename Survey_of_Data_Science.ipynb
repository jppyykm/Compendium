##Linear Regression

Linear regression is the most intuitive exercise of regression, it attempts to model the relationship between two variables by fitting a linear equation to observed data. A linear regression line has an equation of the form $Y = a + bX,$ where $X$ is the explanatory variable and $Y$ is the dependent variable.



````
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
````

### Basis Function Regression

One trick you can use to adapt linear regression to nonlinear relationships between variables is to transform the data according to *basis functions*.

The idea is to take our multidimensional linear model:
$$
y = a_0 + a_1 x_1 + a_2 x_2 + a_3 x_3 + \cdots
$$
and build the $x_1, x_2, x_3,$ and so on, from our single-dimensional input $x$.
That is, we let $x_n = f_n(x)$, where $f_n()$ is some function that transforms our data.

For example, if $f_n(x) = x^n$, our model becomes a polynomial regression:
$$
y = a_0 + a_1 x + a_2 x^2 + a_3 x^3 + \cdots
$$
Notice that this is *still a linear model*—the linearity refers to the fact that the coefficients $a_n$ never multiply or divide each other.
What we have effectively done is taken our one-dimensional $x$ values and projected them into a higher dimension, so that a linear fit can fit more complicated relationships between $x$ and $y$.

````
rng = np.random.RandomState(1)
x = 10 * rng.rand(50)
y = 2 * x - 5 + rng.randn(50)
plt.scatter(x, y);
````

````
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept=True)

model.fit(x[:, np.newaxis], y)

xfit = np.linspace(0, 10, 1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(x, y)
plt.plot(xfit, yfit);
````

````
print("Model slope:    ", model.coef_[0])
print("Model intercept:", model.intercept_)
````

##Logistic Regression

A good starting point, logistic regression is best suited when the output labels are binary and you want yes or no answers. We can view logistic regression as a very simple neural network.

The prediction y-hat is:

$$\hat{y} = \sigma(w^{T}+b)$$ where $$\sigma(z)=\frac{1}{1+e^{-z}}$$

In order to have any predictive power a **loss function** is used to penalize bad guesses:

$$\mathcal{L}(\hat{y},y)= -(y*log(\hat{y}) + (1-y)*log(1-\hat{y}))$$

And a **cost function** to evaluate the outcome:

$$J(w,b)=\frac{1}{m} \sum_{i=1}^{m}\mathcal{L}(\hat{y}^{(i)},y^{(i)})$$

The goal is to minimize the cost function: $$ \min_{w,b} J(x,b)$$

>**Advantages:**

Because of its efficient and straightforward nature, doesn't require high computation power, easy to implement, easily interpretable, used widely by data analyst and scientist. Also, it doesn't require scaling of features. Logistic regression provides a probability score for observations.

>**Disadvantages:**

Logistic regression is not able to handle a large number of categorical features/variables. It is vulnerable to overfitting. Also, can't solve the non-linear problem with the logistic regression that is why it requires a transformation of non-linear features. Logistic regression will not perform well with independent variables that are not correlated to the target variable and are very similar or correlated to each other.



#Artificial Neural Networks

**What is a Neural Network?**

A node structure of functions that iterates seeks an optimal mapping between an input and an output.

Bellow we see a visual representation of a four layer deep neural network with four inputs and a single output.

art-neural-network-image002.webp


In this image a set of inputs labeled A-D are fed into the network to three hidden layers colored red, green and blue who lead to a single output. 

Each node, or colored circle, consists of a linear function followed by a non-linear activation function. Together they are called a neuron and they optimize the network's paramenters when training and choose which nodes to activate or not when evaluating a new example. Similar to a real neuron.

##Notation

>Network Architecture

The first thing we can see from the visualization is the **architecture of the network**. This means how it is structured in layers and how many nodes those layers have, it's shape. In this case we have a 4 layered network with nodes per layer of $[5,4,3,1]$.

Red has five nodes, green has four, blue with three and the final single node output layer. The first input layer is not included in the total depth count of a neural network and thus is denoted as the zeroth layer. Notice how each node is connected to every other node in the layer before and the layer after. This is called a *dense neural network*. More sophisticated variations may not be dense.

The numeric representation of the hidden layers is called the weights, denoted by $W$ and the bias term $b$ are the **parameters** of a model. These change and improve as the algorithm progresses, ideally.

The number of hidden layers and hidden units are both <font color='red'>hyperparameters</font>. These are numbers that the modeler chooses to fine tune a model. We will color these in red for the reminder of the text. 

><font color='red'># of Hidden Layers</font>   ($L$): L ∈ [1,$\infty$], determines the depth of the neural network.

><font color='red'># of Hidden Units</font>   ($U$): U ∈ [1,$\infty$], determines the number of nodes in a layer

>**The Input Vector**

The input is the information (a picture, text, weather data) that is fed into the model denoted by the symbol $x$ in the form of a vector. In this case $$x = [A,B,C,D].$$ 

This is an n by one, $(n,1)$ dimensional vector, where $n$ is the number of dimensions, such as an example $$x = [3,2,1,2]$$ with $(4,1)$ dimensions, here $n = 4$. Following from the illustration we can map input A to the first element of $x$, $A = 3$, $B=2$, $C= 1$ and so on.

The input vector $x$ is one of many training example for the neural network. Neural networks require large amounts of training examples, it could be hundreds, thousands, millions or more. In short, many $x$. 

These can be computed simultaneously using vectorization. Keep in mind that this system is iterative, it will contain many iterations going through it again and again.

By stacking these $x$ vectors we can construct a $(n,m)$ matrix $X$ of all training examples. So a stack of $x=[x_1,x_2,...,x_n]$ for $m$ training examples.
$$X =
\begin{pmatrix}
x_1^{(1)} x_1^{(2)} \cdots x_1^{(m-1)} x_1^{(m)} \\
x_2^{(1)} x_2^{(2)} \cdots x_2^{(m-1)} x_2^{(m)} \\
x_3^{(1)} x_3^{(2)} \cdots x_3^{(m-1)} x_3^{(m)} \\
\vdots \\
x_n^{(1)} x_n^{(2)} \cdots x_n^{(m-1)} x_n^{(m)} \\
\end{pmatrix}
 $$

 We will use the subscript to denote the specific item in the vector. In order to keep up with the iterations needed in machine learning we use the superscript $(i)$ to denote the $i^{th}$ training example. 


>**The Hidden Layers**

The colored circles are called the hidden layers ($l$), this is due to the 'behind the scenes' work done when updating the parameters and dealing with only inputs and outputs as a user. 

These hidden layers contain the parameter information that makes the model function. By tweaking this parameters through forward and back propagation as we will see bellow the model trains itself to improve performance.

The set of parameters these layers have are not suprisingly, arranged in matrix form called $ W^{[l]} $ where the superscript $l$ denotes wich layer it corresponds to. 

The letter $W$ comes from the word weights. This is due to the analogy with weight being a force that modifies the behavior of the network. Some nodes will have more weight than others.

##Gradient Descent

This algorithm searches for an optimal point, it repeats the following:

$$w_{n-1}=w_n-\alpha \frac{dJ(w,b)}{dw}$$

$$b_{n-1}=b_n-\alpha\frac{dJ(w,b)}{db}$$

><font color='red'>Alpha</font>   ($\alpha$): learning rate ∈ [0,1], determines the degree the weights get updated.

##Activation Functions


The function who chooses whether to <font color='yellow'>light</font> up the node or not. 

><font color='red'>Choice of Activation Function</font> determines the behavior of the network.

>**Sigmoid:**

$\sigma(z) = \frac{1}{1+e^{-z}}$ Useful for the last activation of binary classification given the output $[0,1]$


>**Tanh:**

$tanh(z)= \frac{e^z-e^{-z}}{e^z+e^{-z}}$ Hyperbolic Tangent. Better than sigmoid as it "centers" the data with its mean at 0 with range $[-1,1]$

>**Relu:** 

$ReLU(z) = max(0,z)$ Rectified Linear Unit. Most common. 

>**Leaky Relu:** 

$LeLU(z) = max(0.01z,z)$ Slight slope on negative values, derivative not 0.

>**SoftMax:** 

$\sigma (z)_i = \frac{e^{z_i}}{\sum^K_{j=1} e^z_j}. \ $ Normalized Exponential Function. Generalization of the logistic function to multiple dimensions. 

````
#PYTHON IMPLEMENTATIONS
def relu(z):
    result = z.copy()
    result[result < 0] = 0
    return result


def softmax(z):
    e_z = np.exp(z)
    sum_e_z = np.sum(e_z)
    return e_z / sum_e_z
````

##Forward Propagation

image.png

The manner in which computation is done on a neural network is analogous to real neurons as there is a forward pass of stimuli with a return, and so there is a forward pass and a backwards pass. 

For the forward pass we have layer $l$ and weights $W^{[l]}$ with bias $b^{[l]}$. 

For **input** we have $$a^{[l-1]}$$ and for **output** we want the activation for the next layer: $$a^{[l]}$$ 

The computation consists of: $$z^{[l]} = W^{[l]} a^{[l-1]} $$ 

The value of $z$ will be cached for use in the back propagation. Then we apply the activation function g: $$a^{[l]}=g^{[l]}(z^{[l]}) $$ Notice how the activation function has a superscript *l* impliying there can be different activation functions on each node.

##Backward Propagation

Fot the backpass we require **input:** $$da^{[l]}$$ and we get **outputs:** $$da^{[l-1]},dW^{[l]},db^{[l]}$$
The computations follow:
\begin{align}
dz^{[l]} &= da^{[l]} * g^{[l]\prime} (z^{[l]}), \\
dW^{[l]} &= dz^{[l]} \cdot a^{[l-1]^T}, \\
db^{[l]} &= dz^{[l]}, \\
da^{[l-1]} &= W^{[l]^T} \cdot dz^{[l]}
\end{align}

The vectorized implementation in Python is:  

\begin{align}
dZ^{[l]} &= dA^{[l]} * g^{[l]\prime}(Z^{[l]}), \\ 
dW^{[l]} &= \frac{1}{m} dZ^{[l]} \cdot A^{[l-1]^T}, \\ 
db^{[l]} &= \frac{1}{m} np.sum(dZ^{[l]}, axis=1, keepdims =True), \\ 
dA^{[l-1]} &= W^{[l]^T} \cdot dZ^{[l]}
\end{align}

>For the final layer $L$ we compute: $$da^{[L]} = -\frac{y}{a}+ \frac{1-y}{1-a}$$

##Bias and Variance
This is a concept that benefits from explanation by illustration. Let's build some intuition with these figures:


image.png

image.png

One interpretation on the terms bias and variance relative to some baseline performance (Usually human performance, also known as Bayes error), is to view variance as the diference between the train and dev sets relative to the baseline. If it is large then it has **large variance**. 

**High bias** is if both train and dev sets differ considerably but similarly from the baseline. 

The case for **high bias and high variance** is when the train set performance is large compared to the baseline and the dev set performance is large compared to the training set. This is the worse scenario. 

It is often stated a tradeoff between bias and variance but lately we have found that by increasing the network size bias can be lowered and by increasing the amount of data the variance goes down both independently break that tradeoff relationship.

## Best Practices

The **first** thing to do is to look for bias by evaluating training data performance. This can be alleviated with a larger network, longer training times, use a different architecture.

The **second** thing to do is to look for variance via the dev set performance. To alleviate this you can get more data, regularization, a different architecture.


##Regularization
The technique used to reduce the errors by fitting the function appropriately on the given training set and avoid overfitting.

><font color=red>Lambda</font> $\lambda$ Regularization hyperparameter.

$$J(w,b)=\frac{1}{m} \sum_{i=1}^{m}\mathcal{L}(\hat{y}^{(i)},y^{(i)}) + \frac{\lambda}{2m} ||w||^2_2 + \frac{\lambda}{2m} b^2$$

The extra bias term $\frac{\lambda}{2m} b^2$ is usually ommited in practice.

###L2 Regularization

This is the Eucledian norm: $$||w||^2_2=\sum^{n_x}_{j=1}w^2_j=w^Tw$$

This is the most common techique used by far.

###L1 Regularization
$$ \frac{\lambda}{2m} \sum^{n_x}_{i=1} |w| = \frac{\lambda}{2m}||w||_1$$

L1 regularization makes w sparse, lowering the amount of information in the model by adding a lot of zeros into the matrix. This can have a compression effect. The benefits of this have proven to be modest.

When adding all,

$$ J(w^{[1]},b^{[1]},\cdots,w^{[L]},b^{[L]})=\frac{1}{m} \sum_{i=1}^{m}\mathcal{L}(\hat{y}^{(i)},y^{(i)}) + \frac{\lambda}{2m} \sum^{L}_{l=1}||w||^2 $$

Frobenius norm: $$||w^{[l]}||^2 = \sum^{n^l}_{i=1} \sum^{n^{l-1}}_{j=1}  (w^{[l]}_{i,j})^2  $$

Now we change the logistic regression algorithm with:    
$$dw^{[l]} = (backprop) + \frac{\lambda}{m} w^{[l]}$$

$$w^{[l]}:= w^{[l]} - \alpha \cdot dw^{[l]}$$


# Recurring Neural Networks

###Vanilla RNNs



This network's architecture uses information from previous states to influence the predictions at each time step. 

The following illustration shows in blue circles each function step and their input and outputs. The superscript bracket notation $^{<>} $ denotes the time step involved. $h$ stands for the hidden states. $x$ is the input and $\hat{y}$ is the prediction.  

image.png

The function used to calculate a hidden state can be expressed in two ways which are equal. This disctinction becomes less arbitrary in it's implementation and when dealing with matrix dimensionality.



\begin{align}
h^{<t>} &= g(W_h[h^{<t-1>},x^{<t>}]+b_h) \\
h^{<t>} &= g(W_{hh}h^{<t-1>} \oplus W_{hx}x^{<t>}   +b_h)
\end{align}


where, $g()$ is the activation function.

$h$ is the hidden state,

$W_{hh}$ is the weights matrix applied to the previous hidden state,

$W_{hx}$ is the weights matrix applied to the $x$ input,

$W_{yh}$ is the weights matrix applied to get $\hat{y}$

$b$ is the bias term. To each, $b_h$ and $b_y$


Note that the training involves changing $W_{hh}$, $W_{hx}$ ,$W_{yh}$, $b_h$ and $b_y$.

For the prediction at each time step,

$$ \hat{y}^{<t>} = g(W_{yh}h^{<t>} + b_y  $$

####Cross Entropy Loss for RNN

In the case of RNNs we have to take into account all the time steps when calculating a cost function. By having K-classes or possibilities then, 

$$ J = -\frac{1}{T} \sum^K_{t=1} \sum^K_{j=1} y_j^{<t>} \log \hat{y}_j^{<t>}$$

Note that we are simply summing over all the time steps and dividing by T, to get the average cost in each time step. Hence, we are just taking an average through time.

Let's look at the implementation in TensorFlow. We will use an embedding size of 128 and a hidden state size of 16.

````
import numpy as np
from numpy import random
from time import perf_counter
````

````
def sigmoid(x): # Sigmoid function
    return 1.0 / (1.0 + np.exp(-x))
````

````
random.seed(10)                 # Random seed, so your results match ours
emb = 128                       # Embedding size
T = 256                         # Number of variables in the sequences
h_dim = 16                      # Hidden state dimension
h_0 = np.zeros((h_dim, 1))      # Initial hidden state
# Random initialization of weights and biases
w1 = random.standard_normal((h_dim, emb+h_dim))
w2 = random.standard_normal((h_dim, emb+h_dim))
w3 = random.standard_normal((h_dim, emb+h_dim))
b1 = random.standard_normal((h_dim, 1))
b2 = random.standard_normal((h_dim, 1))
b3 = random.standard_normal((h_dim, 1))
X = random.standard_normal((T, emb, 1))
weights = [w1, w2, w3, b1, b2, b3]
````

````
def forward_V_RNN(inputs, weights): # Forward propagation for a a single vanilla RNN cell
    x, h_t = inputs

    # weights.
    wh, _, _, bh, _, _ = weights

    # new hidden state
    h_t = np.dot(wh, np.concatenate([h_t, x])) + bh
    h_t = sigmoid(h_t)

    return h_t, h_t
````

````
forward_V_RNN([X[1],h_0], weights)[0]
````

````
def scan(fn, elems, weights, h_0=None): # Forward propagation for RNNs
    h_t = h_0
    ys = []
    for x in elems:
        ### START CODE HERE (1 lINE) ###
        y, h_t = fn([x, h_t], weights)
        ### END CODE HERE ###
        ys.append(y)
    return ys, h_t
````

````
# vanilla RNNs
tic = perf_counter()
ys, h_T = scan(forward_V_RNN, X, weights, h_0)
toc = perf_counter()
RNN_time=(toc-tic)*1000
print (f"It took {RNN_time:.2f}ms to run the forward method for the vanilla RNN.")
````

The following function inputs a word and returns a list of strings with one character deleted sequentially.

````
def delete_character_list(word, verbose=False):
    '''
    Input:
        word: the string/word for which you will generate all possible words 
                in the vocabulary which have 1 missing character
    Output:
        delete_l: a list of all possible strings obtained by deleting 1 character from word
    '''
    
    delete_l = []
    split_l = []
    
    split_l = [(word[:i],word[i:]) for i in range(len(word)+1)]
    delete_l = [word[:i]+word[i+1:] for i in range(len(word))]

    if verbose: print(f"input word {word}, \nsplit_l = {split_l}, \ndelete_l = {delete_l}")

    return delete_l
````

Switch letter. A function that switches two letters in a word:

````
def switch_letter(word, verbose=False):
    ''' 
    Input:
        word: input string
     Output:
        switches: a list of all possible strings with one adjacent charater switched
    ''' 
    
    switch_l = []
    split_l = []

    split_l = [(word[:i],word[i:]) for i in range(len(word))]
    switch_l = [i + j[1] + j[0] + j[2:] for i,j in split_l if len(j) >= 2]      
    
    if verbose: print(f"Input word = {word} \nsplit_l = {split_l} \nswitch_l = {switch_l}") 

    return switch_l
````

The implementation of a dictionary of word counts for a given text is a powerful tool.

````
from collections import Counter

def get_count(word_l):
    '''
    Input:
        word_l: a set of words representing the corpus. 
    Output:
        worid_count_dict: The wordcount dictionary where key is the word and value is its frequency.
    '''
    
    word_count_dict = {} 
    word_count_dic = Counter()
    for word in word_l: word_count_dic[word] += 1
    word_count_dict = dict(word_count_dic) 
    return word_count_dict
````

Now, given the dictionary of word counts, this function computes the probability each word will appear if randomly selected from the input corpus.

````
def get_probs(word_count_dict):
    '''
    Input:
        word_count_dict: The wordcount dictionary where key is the word and value is its frequency.
    Output:
        probs: A dictionary where keys are the words and the values are the probability that a word will occur. 
    '''
    probs = {}
    probs = {k:float(v)/sum(word_count_dict.values()) for (k,v) in word_count_dict.items()}
    
    return probs
````

##Perplexity

This is a common metric to measure language models. It is related to entropy and it measures how random a text is. An arbitrary list of words will have a higher perplexity than a text generated by a human.


$PP(W) = P(s_{1}, s_{2}, s_{3}, ... , s_{m}) ^{(1/m)}$




Where,

W: Test set containing m sentences s

$s_{i}$: is the ith sentence in the set, each ending with < /s >

m: number of all words in in the set W including </s> but not < s >




The lower the perplexity score the better. It is more likely the sentence will sound natural. Good language models have a perplexity score between 20 and 60.
